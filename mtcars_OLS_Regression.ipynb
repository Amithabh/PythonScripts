```python
from rpy2.robjects import r
from rpy2.robjects import pandas2ri
def data(name): return pandas2ri.ri2py(r[name])
# imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# this allows plots to appear directly in the notebook
%matplotlib inline
```

```python
df = data('mtcars')  
df.describe()  
```

### Implementation of the equivalent of dot in lm(y ~., data = ) in R formula:
df_col = df.columns  
pred = ""  
for i in df_col[1:]:  
    pred = pred + i + "+"  
form = "mpg~"+ pred[:40]  
#### This is the standard import if you're using "formula notation" (similar to R)  
import statsmodels.formula.api as smf  

#### create a fitted model in one line
lm = smf.ols(formula=form, data=df).fit()

```python
# this is the standard import if you're using "formula notation" (similar to R)
import statsmodels.formula.api as smf

# create a fitted model in one line
lm = smf.ols(formula='mpg~wt+am+qsec', data=df).fit()
```

```python
lm.summary()
```

```python
X_new = df.iloc[0:1,[5,6,8]]
# use the model to make predictions on a new value
predicted = lm.predict(X_new)
actual = df.iloc[1,0]
pd.DataFrame({'actual':actual, 'predicted':predicted})
```
